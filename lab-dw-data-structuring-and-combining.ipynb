## Challenge 1: Combining & Cleaning Data

In this challenge, we will be working with the customer data from an insurance company, as we did in the two previous labs. The data can be found here:
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv

But this time, we got new data, which can be found in the following 2 CSV files located at the links below.

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv

Note that you'll need to clean and format the new data.

Observation:
- One option is to first combine the three datasets and then apply the cleaning function to the new combined dataset
- Another option would be to read the clean file you saved in the previous lab, and just clean the two new files and concatenate the three clean datasets

```python
#loading the cleaned csv from previous lab as df_1:
import pandas as pd
url_1 = 'cleaned_dataset.csv' 
df_1 = pd.read_csv(url_1)

#loading the new file2 to clean
url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv'
insurance_db = pd.read_csv(url)

def main():
    insurance_db.columns = [col.lower().replace(" ","_") for col in insurance_db.columns]
    db_1 = insurance_db.rename(columns={"st":"state"})
    db_1["gender"] = db_1["gender"].map({"M":"M","F":"F","Male":"M","female":"F","Femal":"F"})
    db_1["state"] = db_1["state"].replace(["AZ", "Cali", "WA"], ["Arizona","California","Washington"])
    db_1["education"] = db_1["education"].replace(["Bachelors"], ["Bachelor"])
    db_1["customer_lifetime_value"] = db_1["customer_lifetime_value"].str.replace("%","")
    db_1["vehicle_class"] = db_1["vehicle_class"].replace(["Sports Car","Luxury SUV", "Luxury Car"],"Luxury")
    db_1["customer_lifetime_value"]= db_1["customer_lifetime_value"].astype(float)
    db_1[["a","b","c"]] = db_1["number_of_open_complaints"].str.split("/",expand=True)
    db_2 = db_1.drop(columns=["a","c","number_of_open_complaints"])
    db_2.rename(columns={"b":"number_of_open_complaints"},inplace=True)
    db_2["number_of_open_complaints"] = db_2["number_of_open_complaints"].astype("Int64")
    db_3 = db_2.drop_duplicates()
    db_3["gender"].fillna(method="bfill",inplace = True)
    mean = db_3["customer_lifetime_value"].mean()
    db_3["customer_lifetime_value"].fillna(mean,inplace=True)
    return db_3

df_2=main()
df_2
```

```python
df_2.dtypes
df_2.isna().sum()
```

```python
#loading file3 to clean
#loading the new file2 to clean
url_3 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv'
insurance_db_3 = pd.read_csv(url_3)
insurance_db_3.head(3)
```

```python
#file 3 has to follow some cleaning steps but surely less than previous two files. 
def main():
    insurance_db_3.columns = [col.lower().replace(" ","_") for col in insurance_db_3.columns]
    insurance_db_3["vehicle_class"] = insurance_db_3["vehicle_class"].replace(["Sports Car","Luxury SUV", "Luxury Car"],"Luxury")
    df_3 = insurance_db_3.drop_duplicates()
    return df_3

df_3 = main()
df_3
```

```python
#concate 3 data set
new_insurance_df = pd.concat([df_1, df_2, df_3], axis=0)
new_insurance_df
```
